```python
import urllib.request
import sys

import json

try:
  ResultBytes = urllib.request.urlopen("https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/S%C3%A3o%20Vicente?unitGroup=metric&key=N2KS42EPCDCU9U38HL2P7ZDXD&contentType=json")

  # Parse the results as JSON
  jsonData = json.load(ResultBytes)
  for data in jsonData:
    print(jsonData[data])

except urllib.error.HTTPError  as e:
  ErrorInfo= e.read().decode()
  print('Error code: ', e.code, ErrorInfo)
  sys.exit()
except  urllib.error.URLError as e:
  ErrorInfo= e.read().decode()
  print('Error code: ', e.code,ErrorInfo)
  sys.exit()
```
